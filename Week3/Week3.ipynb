```python
"""
Left Skewed Distribution: Mean < Median < Mode
Right Skewed Distribution: Mode < Median < Mean < Kasus kita yang ini > 


Agak sulit menentukannya pake ke 2 hal ini, ada baiknya melakukan hal ini apabila
missinga values nya dibawah 5%

No Skew: Mean = Median = Mode < JARANG PISAN > 
"""
```

```python
print(f'Persentase kolom Age Ngilang {df["Age"].isna().sum() / len(df) * 100}')
""" Kita bakalan Lakuin Random, tapi dengan batasan Mean dan std saja, Well kita ndak bisa banyak 
eksplor, tapi itu salah satu best option, Oh iya Age ini kan tipe datanya Number """

def inputeMissingValues(dat,col):
    mean = dat[col].mean()
    std = dat[col].std()
    nullTotal = dat[col].isnull().sum()
    
    random_age = np.random.randint(mean - std, mean + std, nullTotal)
    nanAge = dat[col].copy()
    nanAge[np.isnan(nanAge)] = random_age
    
    return nanAge
df["Age"] = inputeMissingValues(df,"Age")
```

```python
print(f'Persentase kolom Embarked Ngilang {df["Embarked"].isna().sum() / len(df) * 100}')
print("Ini Kategorikal , so kita isi dengan most common aja")
mostCommon = df["Embarked"].describe()[2]
df["Embarked"].fillna(mostCommon,inplace=True)
```

```python
"""Yowes buat fitur baru. FEATURE ENGINEERING EZ"""
deck = {"A" : 1, "B" : 2, "C" : 3,
       "D" : 4, "E":5, "F" : 6,
       "G" : 7, "None" : 8}

df["Cabin"] = df["Cabin"].fillna("None")
df["Cabin"] = df["Cabin"].map(lambda x:re.compile("([a-zA-Z]+)").search(x).group())
df["Deck"] = df["Cabin"].map(deck)
df["Deck"].fillna(0,inplace=True)
df["Deck"] = df["Deck"].astype(int)


df["Cabin"].map(lambda x:re.compile("([a-zA-Z]+)").search(x))
df.drop(["Cabin"],axis=1,inplace=True)
```

```python
"""Kuy Cek kolom nama"""

df["Title"] = df["Name"].str.extract(" ([a-zA-Z]+)\.")
df['Title'] = df['Title'].replace(['Lady', 'Countess', 'Cap', 'Col', 'Don', 'Dr'\
                                  'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
df["Title"] = df["Title"].replace(["Mlle","Ms"],"Miss")
df["Title"] = df["Title"].replace(["Mme"],"Mrs")

titles = {'Mr': 1, 'Miss': 2, 'Mrs': 3, 'Master': 4, 'Rare': 5}

## Convert into categorical Numbers:
df["Title"] = df["Title"].map(titles)
df["Title"].fillna(0,inplace=True)
df["Title"] = df["Title"].astype(int)
df.drop(["Name"],axis=1,inplace=True)
```

```python
"""Kita Akan buat Range pada Kolom Age, ALasan ? Well kita bisa tau oeluarng mereka """
df["Age"] = df["Age"].astype(int)
np.array(sorted(df["Age"].unique()))
```

```python
df.loc[df["Age"] <= 11,"Age" ] = 0
df.loc[ (df["Age"] > 11) & (df["Age"] <= 18),"Age" ] = 1
df.loc[ (df["Age"] > 18) & (df["Age"] <= 22),"Age" ] = 2
df.loc[ (df["Age"] > 22) & (df["Age"] <= 27),"Age" ] = 3
df.loc[ (df["Age"] > 27) & (df["Age"] <= 33),"Age" ] = 4
df.loc[ (df["Age"] > 33) & (df["Age"] <= 40),"Age" ] = 5
df.loc[ (df["Age"] > 40) & (df["Age"] <= 55),"Age" ] = 6
df.loc[ (df["Age"] > 55),"Age" ]= 7
```

```python
"""Ubah kolom Kategorikal jadi diskrit"""
sex = {"male":1,
      "female":0}
embarked = {"Q":3,
           "C":2,
           "S" : 1}
df["Sex"] = df["Sex"].map(sex)
df["Embarked"] = df["Embarked"].map(embarked)
```
