## sebelum kita masuk, kita belajar dulu tipis2 regex

Regex intinya cuman extract Pola yang udah ditentukan oleh kita, udah itu aja fungsinya :)

[CheatSheet Regex](https://www.activestate.com/wp-content/uploads/2020/03/Python-RegEx-Cheatsheet.pdf)


```python
def checkIfAdded(stringcandidate):
    num1 = re.findall("\d+",stringcandidate)
    if(len(num1) > 0):
        return sum(map(int,num1))
    else:
        return num1

preprocessed = []
for indx,dat in enumerate(mainData[:-1]):
    yow = []
    kecamatan = dat[0]
    positif = checkIfAdded(dat[1])
    mati = checkIfAdded(dat[2])
    sembuh = checkIfAdded(dat[3])
    sisa = checkIfAdded(dat[4])

    
    yow.append(kecamatan)
    yow.append(positif)
    yow.append(mati)
    yow.append(sembuh)
    yow.append(sisa)
 
    preprocessed.append(yow)
preprocessed
```
